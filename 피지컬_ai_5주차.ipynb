```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


data = {
    'Study_Hours': [2,5,8,10,1,3,7,9,4,6],
    'Attendance': [80,90,95, 100,70,85,90,98,88,92],
    'Score': [65,80,85,95,55,70,82,90,75,78]
}
df = pd.DataFrame(data)

X = df[['Study_Hours', 'Attendance']]
y = df['Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,
                                                    random_state = 42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


print("--- [테스트 데이터 예측 결과] ---")
for i in range(len(y_test)):
  print(f"실제 점수 : {y_test.iloc[i]}점 -> 모델의 예측 점수 : {y_pred[i]:.2f}점")

print(f"\n모델의 설명력(R2): {r2_score(y_test, y_pred):.2f}")
```
